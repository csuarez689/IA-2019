```
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

**1. SVM Simple –Cancer de Mama**

**Scikit-Learn contiene la librería svm con muchas clases de algoritmos SVM. Ya que vamos a realizar una tarea de clasificación, usaremos la clase SVC (support vector classifier).**

**Esta clase toma un parámetro que es el tipo de Kernel. Este es muy importante. En el caso de SVM simple setearemos este parámetro a “linear” ya que  SVMs puedo solo clasificar datos separables linealmente.**

**a.- Evalue el modelo usando la matriz de confusión, precisión, recall y F1. Recuerde que dentro de metrics puede obtener las funciones que necesita.**

```
cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target
xTrain, xTest, yTrain, yTest = train_test_split(X,Y,stratify=Y,test_size=0.37)
```

```
model = SVC(kernel='linear')
model.fit(xTrain,yTrain)
pred= model.predict(xTest)

print("Training Accuracy: {:.2f}".format(model.score(xTrain, yTrain)))
print("Test Accuracy: {:.2f}".format(model.score(xTest, yTest)))
print("Confusion Matrix")
mat = confusion_matrix(yTest, pred)
print(mat)
print("\n\nClassification Report")
print(classification_report(yTest,pred))
```

RTA: Se observa una buena proportcion de precision y recall. La precision obtenida indica que el modelo es bueno a la hora de detectar correctamente los casos de cancer.

**b.- Verifique cómo se comporta el clasificador con distintos parámetros de C.¿Por qué piensa que se dio el mejor valor de C? ¿Puede deberse a cierta característica de los datos?**

```
for i in [0.1,1,5,10,100,300,500,800,1000]:
    print()
    model = SVC(kernel='linear',C=i)
    model.fit(xTrain,yTrain)
    pred = model.predict(xTest)
    print("Valor de C= ",i)
    print("Training Accuracy: {:.2f}".format(model.score(xTrain, yTrain)))
    print("Test Accuracy: {:.2f}".format(model.score(xTest, yTest)))
    print()
```

Segun el acurracy el modelo parece mejorar al incrementar la penalizacion por error (C). Pero observando detenidamente se puede ver que el modelo esta sobre sobre ajustando los datos de entrenamiento lo que nos va a dar mayor taza de erores al predecir.
Se puede deber tambien a que algunos datos predominan sobre otros por las medidas de distancias. Quizar lo correcto seria preprocesar los datos para mejorar el procesamiento de los datos en el modelo y asi obtener mejores resultados.
