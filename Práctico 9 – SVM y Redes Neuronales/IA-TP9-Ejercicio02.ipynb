**2.- Realice una comparación entre los resultados de aplicar Random Forest y SVM (RBF) al dataset Iris. Pruebe variando la cantidad de árboles (RF) y C y gamma (SVM) usando GridSearchCV.**
**Antes de comenzar el ejercicio averigue como resuelve la clase SVC los problemas multiclases. ¿Usa la estrategia OVA o OVO?**

```
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```
iris = load_iris()
X=iris.data
Y=iris.target
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size=0.37)
```

**RANDOM FOREST**

---



```
parameters = [{'n_estimators':[10,50,100,150,200],'random_state':[0]}]
rf = RandomForestClassifier()
grid_search = GridSearchCV(estimator = rf, param_grid = parameters, cv= 5, n_jobs = -1, scoring = 'accuracy')
grid_search.fit(xTrain,yTrain)
print("Best Parameters: ", grid_search.best_params_)
print("Best Scoring: {:.2f}".format(grid_search.best_score_))
print("Training accuracy: {:.2f}".format(grid_search.score(xTrain,yTrain)))
print("Test accuracy: {:.2f}".format(grid_search.score(xTest,yTest)))
```

**SVM(RBF)**

---



```
parameters = [{'gamma':[0.1,0.5,2,5,10,50,100,150],
               'C':[0.5,1,2,5,10,50,100,150,200],
               'kernel':['rbf']}]
svc = SVC()
grid_search = GridSearchCV(estimator = svc, param_grid = parameters, cv= 5, n_jobs = -1, scoring = 'accuracy')
grid_search.fit(xTrain,yTrain)
print("Best Parameters: ", grid_search.best_params_)
print("Best Scoring: {:.2f}".format(grid_search.best_score_))
print("Training accuracy: {:.2f}".format(grid_search.score(xTrain,yTrain)))
print("Test accuracy: {:.2f}".format(grid_search.score(xTest,yTest)))
print("Predicciones")
print(grid_search.predict(xTest))
```

SVC usa la estrategia OVR(One-vs-Rest) para resolver problemas multiclases, pero tambien es posible seleccionar la estrategia OVO(One-vs-One).

**a.- Agregue a la comparación solicitada la evaluación de ejecutar sobre el dataset el SVM multiclase con la estrategia que no viene por defecto.**

```
parameters = [{'gamma':[0.1,0.5,2,5,10,50,100,150],
               'C':[0.5,1,2,5,10,50,100,150,200],
               'kernel':['rbf'],
               'decision_function_shape':['ovo']}]
svc = SVC()
grid_search = GridSearchCV(estimator = svc, param_grid = parameters, cv= 5, n_jobs = -1, scoring = 'accuracy')
grid_search.fit(xTrain,yTrain)
print("Best Parameters: ", grid_search.best_params_)
print("Best Scoring: {:.2f}".format(grid_search.best_score_))
print("Training accuracy: {:.2f}".format(grid_search.score(xTrain,yTrain)))
print("Test accuracy: {:.2f}".format(grid_search.score(xTest,yTest)))
print("Predicciones")
print(grid_search.predict(xTest))
```
